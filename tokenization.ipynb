```python
import spacy
from datasets import load_dataset

# Descargar el modelo de idioma español de Spacy
nlp = spacy.load("es_core_news_sm")

```

```python

dataset = load_dataset("large_spanish_corpus", "DGT", split="train")
```

```python
#Verificar que sí trajo las sentencias
sentences = dataset["text"]
dataset["text"][0]
```

```python

import re

# Definir la expresión regular - solo tokens con letras
pattern = r'^[a-zA-Z]+$'
# Función para tokenizar y limpiar una sentencia
def tokenize(sentence):
    # Tokenizar la sentencia con Spacy
    doc = nlp(sentence)

    # Filtrar los tokens que no son stopwords, números, ni puntuación
    sentence_tok = [token.text.lower() for token in doc if  not token.is_stop and re.match(pattern, token.text) ] #

    return sentence_tok

tokenize("El Ejemplo es  dificiL de probar 20/21 20.11.2007")
```

```python
# TOKENIZAR SOLO LAS PRIMERAS 20 SENTENCIAS
sentences_tokenized = []

def tokenize_section_of_dataset(start, end):

  for idx in range(start,end):
      # Agregar la sentencia a la lista de sentencias tokenizadas
       sentences_tokenized.append(tokenize(sentences[idx]))

tokenize_section_of_dataset(0,20)



# Mostrar las primeras 20 sentencias tokenizadas
for i, sentence in enumerate(sentences_tokenized[:20], start=1):
    print(sentence)

```

```python
#tokenizar TODAS las sentencias

sentences_tokenized = []

def tokenize_all_sentences():
    for sentence in sentences:
        # Agregar la sentencia a la lista de sentencias tokenizadas
        sentences_tokenized.append(tokenize(sentence))

tokenize_all_sentences()

print(f"Número de sentencias tokenizadas: {len(sentences_tokenized)}")

# Mostrar las primeras 20 sentencias tokenizadas
for i, sentence in enumerate(sentences_tokenized, start=1):
    print(sentence)
```

```python
#Guardar las tokenizadas en un .json
import json

# Nombre del archivo JSON
tokenizadas_json = "tokenizadas.json"

# Guardar la lista en el archivo JSON
with open(tokenizadas_json, "w") as archivo:
    json.dump(sentences_tokenized, archivo)

print("Lista guardada en", tokenizadas_json)
# Cuantas se tokenizaron
len(sentences_tokenized)
```
